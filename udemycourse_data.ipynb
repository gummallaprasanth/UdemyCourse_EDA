```python
# 10. Which Course Has More Lectures (Free or Paid)?
# by using groupby method

df.groupby('is_paid').count()["course_title"]
```

```python
#  Which Course Has More Lectures (Free or Paid)?

# disply the all columns means

df.groupby(['is_paid']).mean(numeric_only=True)
```

```python
# printing the required columns for the better undestanding
# 11. Which Courses Have A Higher Number of Subscribers Free or Paid?


df.groupby(['is_paid']).mean(numeric_only=True)['num_subscribers']
```

```python

# 12. Which Level Has The Highest Number of Subscribers?

df.groupby(['level']).mean(numeric_only=True)
```

```python
# 12. Which Level Has The Highest Number of Subscribers?(using barplots)

sns.barplot(x=df['level'],y=df['num_subscribers'],data=df,hue="level")
plt.show()
```

```python
# 13. Find Most Popular Course Title
# if the channel has more number of subscribers that is the most popular

df[df['num_subscribers']==df['num_subscribers'].max()]['course_title']
```

```python
# 14. Display 10 Most Popular Courses As Per Number of Subscribers

index=df['num_subscribers'].sort_values(ascending=False).head(10).index
```

```python
# . Display 10 Most Popular Courses As Per Number of Subscribers
# using loc method

df.loc[index][['course_title','num_subscribers']]
```

```python
# 15. Find The Course Which Is Having The Highest Number of Reviews.

df[df['num_reviews'].max()==df['num_reviews']][['course_title','num_reviews']]
```

```python
#  Display 10 Most Popular Courses As Per Number of Subscribers
# using the seaborn barplots for better visualizing


top_10=df.sort_values(by='num_subscribers',ascending=False).head(10)
```

```python
sns.barplot(x="num_subscribers",y="course_title",data=top_10,hue="num_subscribers",palette="viridis")
```

```python
# 16. Does Price Affect the Number of Reviews?
# ploting the scatter plot to show the relation between price and num_reviews

plt.figure(figsize=(10,6))
sns.scatterplot(x='price',y='num_reviews',data=df)
```
